K-means is an unsupervised learning method for clustering data points. The algorithm iteratively divides data points into K clusters by minimizing the variance in each cluster.

First, each data point is randomly assigned to one of the K clusters. Then, we compute the centroid of each cluster, and reassign each data point to the cluster with the closest centroid. We repeat this process until the cluster assignments for each data point remain the same. 

```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show()
```

We use the Elbow method to determine at which point in the graph gives us the optimal value for K. Below we see that K = 2 is a good value.

```python
from sklearn.cluster import KMeans

data = list(zip(x, y))
inertias = []

for i in range(1, 11): # we have 10 data points, so maximum number of clusters is 10
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_) # inertia_ is the sum of squared distances of samples to their closest cluster center

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
