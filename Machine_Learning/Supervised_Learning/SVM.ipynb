```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
X, y = make_blobs(n_samples=100, centers=2, random_state=42)

# Create a support vector machine classifier
clf = SVC(kernel='linear', C=1)

# Train the classifier on the training data
clf.fit(X, y)

# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')

# Create a mesh to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Plot the decision boundary
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.8)

plt.show()
```

In this example, we use the make_blobs function from sklearn.datasets to generate sample data with two clusters, then we create a SVC classifier object with a linear kernel and C=1 as the regularization parameter, we use the fit method to train the classifier on the data. Finally, we use matplotlib to plot the data points and the decision boundary, the decision boundary is obtained by creating a mesh of points in the range of the data and then using the predict method to obtain the class of each point of this mesh.

This will show a 2-D plot with the data points, the color of the points represents the class they belong to, and the background color represents the class predicted by the SVM model.
